```python
#cau 14
import numpy as np
#Nhập số dòng và cột của ma trận:
n=int(input("n="))
m=int(input("m="))
def myfunc(n,m):
  data = np.random.randint(0, 100, size=(n, m))
  print(data)
  print("**************")
  k=int(input("Nhập k:"))
  print("dòng k",data[k])
  print("cột k",data[:,k:k+1])
  max=np.sum(data[0])
  for i in range(n):
     row=data[i]
     if(np.sum(row)>max):
      max=np.sum(row)
  print("dòng có tổng lớn nhất:",max)
  min=np.prod(data[:,:1])
  for i in range(m):
    column=data[:,:i+1]
    if(np.prod(column)<min):
      min=np.prod(column)
  print("Cột có tích nhỏ nhất:",min)
  # In ra các phần tử thuộc dòng chẵn và cột lẻ
  print("Các phần tử ở dòng chẵn và cột lẻ:")
  found_elements = []
  for i in range(n): # Duyệt qua chỉ số dòng
    if i % 2 == 0: # Kiểm tra dòng chẵn (chỉ số 0, 2, 4,...)
      for j in range(m): # Duyệt qua chỉ số cột
        if j % 2 != 0: # Kiểm tra cột lẻ (chỉ số 1, 3, 5,...)
          found_elements.append(data[i, j])
          print(f"({i}, {j}): {data[i, j]}")
  if not found_elements:
      print("Không tìm thấy phần tử nào thỏa mãn điều kiện.")
  print(f"Giá trị trung bình các phần tử thuộc biên: {(np.sum(data[:,:1])+np.sum(data[:,m-1:]))/(2*m)}")
  size_data=np.size(data)
 # print(data[:,1:m-1])
  print(f"Giá trị trung bình tích các phần tử khôngthuộc biên: {np.prod(data[:,1:m-1])**(1/(size_data-n*2))}")
myfunc(n,m)
```

```python
#Câu 15:
import datetime

# a. Xây dựng cấu trúc SINHVIEN mô tả một sinh viên
class SINHVIEN:
    def __init__(self, ma_sinh_vien, ten, nam_sinh, diem_trung_binh):
        # Validate ma_sinh_vien (10 ký tự)
        if not isinstance(ma_sinh_vien, str) or len(ma_sinh_vien) != 10:
            raise ValueError("Mã sinh viên phải là chuỗi có 10 ký tự.")
        self.ma_sinh_vien = ma_sinh_vien

        # Validate ten (tối đa 20 ký tự)
        if not isinstance(ten, str) or len(ten) > 20:
            raise ValueError("Tên phải là chuỗi tối đa 20 ký tự.")
        self.ten = ten

        # Validate nam_sinh (số nguyên)
        if not isinstance(nam_sinh, int) or not (1900 <= nam_sinh <= datetime.datetime.now().year):
            raise ValueError("Năm sinh phải là số nguyên hợp lệ.")
        self.nam_sinh = nam_sinh

        # Validate diem_trung_binh (số thực)
        if not isinstance(diem_trung_binh, (int, float)) or not (0 <= diem_trung_binh <= 10):
            raise ValueError("Điểm trung bình phải là số thực từ 0 đến 10.")
        self.diem_trung_binh = diem_trung_binh

    def __str__(self):
        return f"Mã SV: {self.ma_sinh_vien}, Tên: {self.ten}, Năm sinh: {self.nam_sinh}, ĐTB: {self.diem_trung_binh}"

# b. Cho một mảng có n sinh viên.
# Ví dụ tạo một mảng (list) các sinh viên
students = [
    SINHVIEN("01TH000001", "Nguyen Van A", 2003, 7.5),
    SINHVIEN("02DH000002", "Tran Thi B", 2004, 4.8),
    SINHVIEN("03TH000003", "Le Van C", 2002, 9.0),
    SINHVIEN("04DH000004", "Pham Thi Lan", 2005, 6.2),
    SINHVIEN("05TH000005", "Hoang Van D", 2001, 3.5),
    SINHVIEN("06DH000006", "Phan Thi E", 2004, 5.0),
    SINHVIEN("07KT000007", "Nguyen Thi Lan", 2003, 8.1)
]

print("Danh sách sinh viên:")
for s in students:
    print(s)
```

```python
# c. Viết hàm cho biết có bao sinh viên đủ điều kiện lên lớp
def count_eligible_students(student_list):
    count = 0
    for student in student_list:
        if student.diem_trung_binh >= 5.0:
            count += 1
    return count

eligible_count = count_eligible_students(students)
print(f"\nSố sinh viên đủ điều kiện lên lớp: {eligible_count}")
```

```python
import datetime

# d. Xuất các sinh viên đủ 20 tuổi
def print_20_year_old_students(student_list):
    current_year = datetime.datetime.now().year
    print("\nCác sinh viên đủ 20 tuổi:")
    found = False
    for student in student_list:
        if (current_year - student.nam_sinh) == 20:
            print(student)
            found = True
    if not found:
        print("Không có sinh viên nào đủ 20 tuổi.")

print_20_year_old_students(students)
```

```python
# e. Đếm số sinh viên học hệ đại học
def count_university_students(student_list):
    count = 0
    for student in student_list:
        # Mã sinh viên chứa 2 ký tự DH ở vị trí 2,3 (index 1, 2)
        if student.ma_sinh_vien[1:3] == "DH":
            count += 1
    return count

university_students_count = count_university_students(students)
print(f"\nSố sinh viên học hệ đại học: {university_students_count}")
```

```python
# f. Cho biết trong mảng có bao nhiêu sinh viên có tên «Lan»
def count_students_by_first_name(student_list, first_name):
    count = 0
    for student in student_list:
        # Tách tên cuối cùng
        student_first_name = student.ten.split()[-1]
        if student_first_name == first_name:
            count += 1
    return count

lan_count = count_students_by_first_name(students, "Lan")
print(f"\nSố sinh viên có tên 'Lan': {lan_count}")
```

```python
# g. Cho biết trong mảng có bao nhiêu sinh viên có họ «Phan»
def count_students_by_last_name(student_list, last_name):
    count = 0
    for student in student_list:
        # Tách họ
        student_last_name = student.ten.split()[0]
        if student_last_name == last_name:
            count += 1
    return count

phan_count = count_students_by_last_name(students, "Phan")
print(f"\nSố sinh viên có họ 'Phan': {phan_count}")
```
