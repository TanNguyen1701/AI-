```python
#BÀI 3
import math

# Cài đặt biến toàn cục
N = 5             # Kích thước bàn cờ NxN (Có thể đổi thành 10)
WIN_CONDITION = 4 # Số quân liên tiếp cần để thắng (Caro 5x5 thường đánh 4 quân thắng)
X = "X"           # Máy
O = "O"           # Người chơi
EMPTY = " "

def print_board(board):
    print("\n   " + "   ".join([str(i) for i in range(N)]))
    for i, row in enumerate(board):
        print(f"{i}  " + " | ".join(row))
        print("   " + "-" * (4 * N - 1))

def check_win_nxn(board, player):
    directions = [(0, 1), (1, 0), (1, 1), (1, -1)]
    for r in range(N):
        for c in range(N):
            if board[r][c] == player:
                for dr, dc in directions:
                    count = 0
                    for i in range(WIN_CONDITION):
                        nr, nc = r + dr * i, c + dc * i
                        if 0 <= nr < N and 0 <= nc < N and board[nr][nc] == player:
                            count += 1
                        else: break
                    if count == WIN_CONDITION: return True
    return False

def is_full(board):
    return all(cell != EMPTY for row in board for cell in row)

# Hàm lượng giá sơ bộ
def evaluate_board(board):
    if check_win_nxn(board, X): return 100
    if check_win_nxn(board, O): return -100
    return 0

# THUẬT TOÁN MINIMAX (KHÔNG CẮT TỈA)
def minimax_nxn(board, depth, is_maximizing):
    score = evaluate_board(board)
    if score == 100 or score == -100: return score
    if depth == 0 or is_full(board): return score # Giới hạn độ sâu

    if is_maximizing:
        max_eval = -math.inf
        for i in range(N):
            for j in range(N):
                if board[i][j] == EMPTY:
                    board[i][j] = X
                    eval = minimax_nxn(board, depth - 1, False)
                    board[i][j] = EMPTY
                    max_eval = max(max_eval, eval)
        return max_eval
    else:
        min_eval = math.inf
        for i in range(N):
            for j in range(N):
                if board[i][j] == EMPTY:
                    board[i][j] = O
                    eval = minimax_nxn(board, depth - 1, True)
                    board[i][j] = EMPTY
                    min_eval = min(min_eval, eval)
        return min_eval

def best_move_minimax(board):
    best_val = -math.inf
    move = (-1, -1)
    # Giới hạn độ sâu = 2 cho Minimax vì không có cắt tỉa sẽ chạy rất chậm
    for i in range(N):
        for j in range(N):
            if board[i][j] == EMPTY:
                board[i][j] = X
                move_val = minimax_nxn(board, 2, False)
                board[i][j] = EMPTY
                if move_val > best_val:
                    best_val = move_val
                    move = (i, j)
    # Nếu bảng trống, đánh vào giữa
    if move == (-1, -1) or best_val == -math.inf:
        return (N//2, N//2)
    return move

# --- GIAO DIỆN CHƠI TRÊN TERMINAL ---
if __name__ == "__main__":
    board = [[EMPTY for _ in range(N)] for _ in range(N)]
    print(f"BÀI 3: TIC-TAC-TOE {N}x{N} - MINIMAX (CẦN {WIN_CONDITION} QUÂN ĐỂ THẮNG)")
    print_board(board)

    while True:
        while True:
            try:
                r, c = map(int, input("Nhập tọa độ nước đi (hàng cột, ví dụ: 2 3): ").split())
                if 0 <= r < N and 0 <= c < N and board[r][c] == EMPTY:
                    board[r][c] = O
                    break
                else: print("Nước đi không hợp lệ!")
            except: print("Vui lòng nhập đúng định dạng.")

        print_board(board)
        if check_win_nxn(board, O):
            print("Bạn thắng!"); break
        if is_full(board):
            print("Hòa!"); break

        print("Máy đang tính toán (Minimax)...")
        r, c = best_move_minimax(board)
        board[r][c] = X
        print_board(board)

        if check_win_nxn(board, X):
            print("Máy thắng!"); break
        if is_full(board):
            print("Hòa!"); break
```

```python
#BÀI 4
import math

# Các thiết lập kích thước và điều kiện thắng giống hệt Bài 3
N = 5
WIN_CONDITION = 4
X, O, EMPTY = "X", "O", " "

# (Giữ nguyên các hàm print_board, check_win_nxn, is_full, evaluate_board từ Bài 3)
def print_board(board):
    print("\n   " + "   ".join([str(i) for i in range(N)]))
    for i, row in enumerate(board):
        print(f"{i}  " + " | ".join(row))
        print("   " + "-" * (4 * N - 1))

def check_win_nxn(board, player):
    directions = [(0, 1), (1, 0), (1, 1), (1, -1)]
    for r in range(N):
        for c in range(N):
            if board[r][c] == player:
                for dr, dc in directions:
                    count = 0
                    for i in range(WIN_CONDITION):
                        nr, nc = r + dr * i, c + dc * i
                        if 0 <= nr < N and 0 <= nc < N and board[nr][nc] == player: count += 1
                        else: break
                    if count == WIN_CONDITION: return True
    return False

def is_full(board):
    return all(cell != EMPTY for row in board for cell in row)

def evaluate_board(board):
    if check_win_nxn(board, X): return 100
    if check_win_nxn(board, O): return -100
    return 0

# THUẬT TOÁN ALPHA-BETA CHO NxN
def alphabeta_nxn(board, depth, alpha, beta, is_maximizing):
    score = evaluate_board(board)
    if score == 100 or score == -100: return score
    if depth == 0 or is_full(board): return score

    if is_maximizing:
        max_eval = -math.inf
        for i in range(N):
            for j in range(N):
                if board[i][j] == EMPTY:
                    board[i][j] = X
                    eval = alphabeta_nxn(board, depth - 1, alpha, beta, False)
                    board[i][j] = EMPTY
                    max_eval = max(max_eval, eval)
                    alpha = max(alpha, eval)
                    if beta <= alpha: break # Cắt tỉa
        return max_eval
    else:
        min_eval = math.inf
        for i in range(N):
            for j in range(N):
                if board[i][j] == EMPTY:
                    board[i][j] = O
                    eval = alphabeta_nxn(board, depth - 1, alpha, beta, True)
                    board[i][j] = EMPTY
                    min_eval = min(min_eval, eval)
                    beta = min(beta, eval)
                    if beta <= alpha: break # Cắt tỉa
        return min_eval

def best_move_alphabeta(board):
    best_val = -math.inf
    move = (-1, -1)
    # Alpha-Beta nhanh hơn nên có thể duyệt tới Depth = 3
    for i in range(N):
        for j in range(N):
            if board[i][j] == EMPTY:
                board[i][j] = X
                move_val = alphabeta_nxn(board, 3, -math.inf, math.inf, False)
                board[i][j] = EMPTY
                if move_val > best_val:
                    best_val = move_val
                    move = (i, j)
    if move == (-1, -1) or best_val == -math.inf:
        return (N//2, N//2)
    return move

# --- GIAO DIỆN CHƠI TRÊN TERMINAL ---
if __name__ == "__main__":
    board = [[EMPTY for _ in range(N)] for _ in range(N)]
    print(f"BÀI 4: TIC-TAC-TOE {N}x{N} - ALPHA-BETA PRUNING")
    print_board(board)

    while True:
        while True:
            try:
                r, c = map(int, input("Nhập tọa độ nước đi (hàng cột): ").split())
                if 0 <= r < N and 0 <= c < N and board[r][c] == EMPTY:
                    board[r][c] = O
                    break
                else: print("Lỗi: Nước đi trùng hoặc ngoài bàn cờ!")
            except: print("Lỗi: Định dạng không đúng.")

        print_board(board)
        if check_win_nxn(board, O):
            print("Bạn thắng!"); break
        if is_full(board):
            print("Hòa!"); break

        print("Máy đang tính toán bằng Alpha-Beta...")
        r, c = best_move_alphabeta(board)
        board[r][c] = X
        print_board(board)

        if check_win_nxn(board, X):
            print("Máy thắng!"); break
        if is_full(board):
            print("Hòa!"); break
```
