```python
#BÀI 1&2
import math

X = "X"
O = "O"
EMPTY = " "

def print_board(board):
    print("\n")
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

def check_winner(board):
    # Kiểm tra hàng và cột
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != EMPTY: return board[i][0]
        if board[0][i] == board[1][i] == board[2][i] != EMPTY: return board[0][i]
    # Kiểm tra chéo
    if board[0][0] == board[1][1] == board[2][2] != EMPTY: return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != EMPTY: return board[0][2]
    return None

def is_full(board):
    return all(cell != EMPTY for row in board for cell in row)

# Thuật toán Alpha-Beta (Kết hợp Minimax)
def minimax_alpha_beta(board, depth, alpha, beta, is_maximizing):
    winner = check_winner(board)
    if winner == X: return 10 - depth # Máy (X) thắng
    if winner == O: return -10 + depth # Người (O) thắng
    if is_full(board): return 0

    if is_maximizing:
        max_eval = -math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == EMPTY:
                    board[i][j] = X
                    eval = minimax_alpha_beta(board, depth + 1, alpha, beta, False)
                    board[i][j] = EMPTY
                    max_eval = max(max_eval, eval)
                    alpha = max(alpha, eval)
                    if beta <= alpha: break # Cắt tỉa Alpha-Beta
        return max_eval
    else:
        min_eval = math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == EMPTY:
                    board[i][j] = O
                    eval = minimax_alpha_beta(board, depth + 1, alpha, beta, True)
                    board[i][j] = EMPTY
                    min_eval = min(min_eval, eval)
                    beta = min(beta, eval)
                    if beta <= alpha: break # Cắt tỉa Alpha-Beta
        return min_eval

def best_move(board):
    best_val = -math.inf
    move = (-1, -1)
    for i in range(3):
        for j in range(3):
            if board[i][j] == EMPTY:
                board[i][j] = X
                move_val = minimax_alpha_beta(board, 0, -math.inf, math.inf, False)
                board[i][j] = EMPTY
                if move_val > best_val:
                    best_val = move_val
                    move = (i, j)
    return move

# Vòng lặp trò chơi
board = [[EMPTY for _ in range(3)] for _ in range(3)]
print("Tic-Tac-Toe 3x3 - Bạn là O, Máy là X")
print_board(board)

while True:
    # Người chơi đi (O)
    while True:
        try:
            r, c = map(int, input("Nhập nước đi của bạn (hàng cột, ví dụ: 0 1): ").split())
            if board[r][c] == EMPTY:
                board[r][c] = O
                break
            else: print("Ô này đã có người đánh!")
        except: print("Nhập không hợp lệ, vui lòng nhập lại.")

    print_board(board)
    if check_winner(board) or is_full(board): break

    # Máy đi (X)
    print("Máy đang suy nghĩ...")
    r, c = best_move(board)
    board[r][c] = X
    print_board(board)

    if check_winner(board) or is_full(board): break

winner = check_winner(board)
if winner: print(f"Người chiến thắng: {winner}")
else: print("Hòa!")
```
